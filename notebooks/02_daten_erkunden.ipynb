```python
import pandas as pd\n
import matplotlib.pyplot as plt\n
\n
df = pd.read_csv('daten/haeuser.csv')\n
df.head()
```

```python
print('Form:', df.shape)\n
print('Spalten:', list(df.columns))\n
print('Fehlende Werte:')\n
print(df.isna().sum())
```

```python
display(df.describe())
```

```python
plt.figure(figsize=(6,4))\n
plt.hist(df['preis_euro'], bins=15, color='steelblue', edgecolor='black')\n
plt.title('Verteilung der Hauspreise')\n
plt.xlabel('Preis (Euro)')\n
plt.ylabel('Anzahl')\n
plt.tight_layout()\n
plt.show()
```

```python
plt.figure(figsize=(6,4))\n
plt.scatter(df['groesse_m2'], df['preis_euro'], alpha=0.7, color='darkorange')\n
plt.title('Groesse vs. Preis')\n
plt.xlabel('Groesse (m2)')\n
plt.ylabel('Preis (Euro)')\n
plt.tight_layout()\n
plt.show()
```
